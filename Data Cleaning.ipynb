```python
import pandas as pd
```

```python
df= pd.read_csv("social_media_engagement_dataset.csv")
```

```python
df.isnull().sum()
```

```python
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"Duplicate rows removed: {before - after}")
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
print("Timestamp dtype:", df['Timestamp'].dtype)
```

```python
# Total interactions column
df['Total_Interactions'] = df[['Likes', 'Comments', 'Shares', 'Saves']].sum(axis=1)

# Hashtag bucket
df['Hashtag_Bucket'] = pd.cut(
    df['Hashtag_Count'],
    bins=[-1, 5, 10, df['Hashtag_Count'].max()],
    labels=['Low', 'Medium', 'High']
)

df.head()
```

```python
print("Final shape:", df.shape)
print("Remaining nulls:", df.isnull().sum().sum())
print("Columns:", list(df.columns))
```

```python
df.to_csv('social_media_engagement_clean.csv', index=False)
print("Saved: social_media_engagement_clean.csv")
```
